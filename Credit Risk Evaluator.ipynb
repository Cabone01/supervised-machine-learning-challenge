```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
```

```python
# Import the data
df = pd.read_csv('Resources/lending_data.csv')
df.head()
```

## Predicting which is better before running test
I believe the Random Forest Classifier will perform better. The reason being is that Random Forest Classifier I believe performs better when more variables are introduced to the dataset that could affect the performance. It will also break it apart further with decisions tress allowing for more true or false tests. With Logistic Regression, it tries to create the best line model for the results to distinguish them.

```python
y = df['loan_status'].values
X = df.drop('loan_status', axis=1)
X.head()
```

```python
# Split the data into X_train, X_test, y_train, y_test
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
X_train.head()
```

```python
# Train a Logistic Regression model print the model score
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(max_iter=10000)
classifier.fit(X_train, y_train)
print(f"Training Model Score: {classifier.score(X_train, y_train)}")
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier

scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
clf = RandomForestClassifier(random_state=42, n_estimators=500).fit(X_train_scaled, y_train)
print(f'Training Model Score: {clf.score(X_train_scaled, y_train)}')
```

## Results
My beliefs were correct despite how minimual it may appear. The Random Forest performed better by being only 0.006 better than then the Logistic Regression model. However even though it did slightly better, it is a small enough difference where either can serve the same purpose without worry. The benefit of using logistic or random forest is that it is able to be trained faster which I noticed while running the code. Perhaps if the data was larger then the difference would be seen more but for cases like these either can work.
